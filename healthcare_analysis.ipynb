```python
df["Medical Condition"].value_counts().plot(kind="bar", figsize=(8,5))
plt.title("Patients by Medical Condition")
plt.xlabel("Medical Condition")
plt.ylabel("Number of Patients")
plt.xticks(rotation=45)
plt.show()
```

```python
df["Gender"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

```python
df["Admission Type"].value_counts().plot(kind="bar")
plt.title("Admission Type Distribution")
plt.xlabel("Admission Type")
plt.ylabel("Count")
plt.show()
```

```python
df["Age"].plot(kind="hist", bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
```

```python
df["Billing Amount"].plot(kind="hist", bins=20)
plt.title("Billing Amount Distribution")
plt.xlabel("Billing Amount")
plt.show()
```

```python
print("Average Age:", df["Age"].mean())
print("Average Billing Amount:", df["Billing Amount"].mean())
print("Most Common Medical Condition:", df["Medical Condition"].mode()[0])
print("Most Common Admission Type:", df["Admission Type"].mode()[0])
```
